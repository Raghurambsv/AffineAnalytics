import seaborn as sns

%matplotlib inline

# Dataset

```python
from sklearn.datasets import make_blobs

data, labels = make_blobs(n_samples=100, n_features=4, centers=4,cluster_std=4,random_state=2)
```

# visualization

```python
plt.scatter(data[:,0], data[:,1], c = labels, cmap='coolwarm');
```
