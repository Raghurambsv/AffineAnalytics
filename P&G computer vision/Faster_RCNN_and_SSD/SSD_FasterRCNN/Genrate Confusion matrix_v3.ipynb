## Combining the ground truth and predictions

```python
df_actual.columns = [c+'_actual' if c not in ["Image_key","Image_path"] else c for c in df_actual.columns]
```

```python
df_conf_mat = df_actual.merge(df_after_infer)
```

```python
df_conf_mat.to_csv("QC_iter7_cutoff0.4_test.csv")
```

```python
defect_cols = [c for c in df_after_infer.columns if c not in ['Image_key','Unknown','Good']]
```

## Getting the confusion Matrix

```python
for c in defect_cols:
    req_cols = [col for col in df_conf_mat.columns if col.startswith(c)]
    print(req_cols)
    df_conf_mat[c+'_TP'] = df_conf_mat[req_cols]\
        .apply(lambda row: row[req_cols[0]] if row[req_cols[0]]<row[req_cols[-1]] else row[req_cols[-1]],axis=1)
    df_conf_mat[c+'_FP'] = df_conf_mat[req_cols]\
        .apply(lambda row: row[req_cols[-1]]-row[req_cols[0]] if row[req_cols[0]]<row[req_cols[-1]] else 0,axis=1)
    df_conf_mat[c+'_TN'] = df_conf_mat[req_cols]\
        .apply(lambda row: 1 if (row[req_cols[0]]==0 and row[req_cols[-1]]==0) else 0,axis=1)
    df_conf_mat[c+'_FN'] = df_conf_mat[req_cols+[c+'_TP']]\
        .apply(lambda row: (row[req_cols[0]]-row[c+'_TP']),axis=1)
```

```python
req_cols = [c for c in df_conf_mat.columns if c.endswith("actual") or c.endswith("TP") or c.endswith("FP") or c.endswith("TN") or c.endswith("FN")]
```

```python
req_cols
```

## Writing the confusion matrix

```python
df_conf_mat[req_cols].sum().to_csv("confusion_matrix_iter7_cutoff0.4_test.csv")
```
